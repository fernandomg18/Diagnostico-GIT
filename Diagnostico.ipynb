```python
def main(df):
  print("top 10 tweets más retweeted\n")
  top10_retweet(df)
  print("\ntop 10 usuarios en función a la cantidad de tweets que emitieron\n")
  top10_users_tweets(df)
  print("\nLos top 10 días donde hay más tweets\n")
  top10_days(df)
  top10_hashtag(df)
```

```python
def top10_retweet(df):
  by_retweetCount = df.sort_values('retweetCount',ascending=False)
  print(by_retweetCount["content"].head(10))
```

```python
def top10_users_tweets(df):
  df['user'] = df['user'].apply(lambda x: x["username"])
  by_users_tweets = df.groupby(df["user"]).count().sort_values("content", ascending=False)
  print(by_users_tweets["content"].head(10))
```

```python
def top10_days(df):
  df['date'] = df['date'].apply(lambda x: x.date())
  by_days = df.groupby(["date"]).count().sort_values('content',ascending=False)
  print(by_days["content"].head(10))
```
